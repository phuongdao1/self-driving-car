## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image




# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
````

### Step 1: Extract color white and yellow

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images directory.

# image is expected be in RGB color space
def extract_white_and_yellow(image): 
    lower_white = np.uint8([200, 200, 200])
    upper_white = np.uint8([255, 255, 255])
    white_mask = cv2.inRange(image, lower_white, upper_white)

    lower_yellow = np.uint8([175, 175,   0])
    upper_yellow = np.uint8([255, 255, 204])
    yellow_mask = cv2.inRange(image, lower_yellow, upper_yellow)
    
    white_yellow_masked = cv2.bitwise_and(image, image, mask = cv2.bitwise_or(white_mask, yellow_mask))
    
    gray_img = grayscale(white_yellow_masked)
    masked = cv2.bitwise_and(image, image, mask = gray_img)
    
    return masked
    
    
images_after_color_extracted = list(map(extract_white_and_yellow, test_images))
show_processed_images(test_images, images_after_color_extracted)
````

### Step 2: Canny edge mask

````python
def canny_edge_mask(image_after_color_extracted):
    kernel_size = 5
    blurred_img = gaussian_blur(image_after_color_extracted, kernel_size)
    
    canny_img = canny(blurred_img, 30, 100)
    
    w = canny_img.shape[1]
    h = canny_img.shape[0]
    vertices = np.array([[(0.05 * w, h), (w * 0.475, h * 0.58), 
                          (w * 0.525, h * 0.58), (0.95 * w, h)]], dtype=np.int32)
    
    return region_of_interest(canny_img, vertices)
    

images_after_canny_mask = list(map(canny_edge_mask, images_after_color_extracted))
show_processed_images(test_images, images_after_canny_mask)
````

### Step 3: Hough transform and sketch lanes

In this step, after receiving line segments from Hough transformation, we need to detect two average lines corresponding to two lanes from these segments. In order to do that, I first cluster the line segments based on their slopes and intercepts using k-means with k=2. I have used clustering here to take into account of the cases both slopes are negative or positive which can happens in other situations eventhough in our our test cases, one slope is positive and another one is negative.

After clustering to extract the average lane lines, I proceed to remove segment outliers. This is done based on computing the distance of segment endpoints to the average lane lines. After removing all the outliers, I recompute the average lane lines.

````python
from sklearn.cluster import KMeans
import math
import re
from PIL import Image

def dist_to_line(x0, y0, slope, intercept):
    return (abs(slope*x0 - y0 + intercept) / (math.sqrt(slope*slope+1)))

def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    # record slopes and intercepts of all the lines
    line_params = []
    
    # compute slope and intercept for every line
    for line in lines:
        for x1,y1,x2,y2 in line:
            if abs(x2 - x1) > 2:
                slope = (y2 - y1) / (x2 - x1)
                if abs(slope) > 0.5:
                    intercept = (y2 - x2 * slope)
                    line_params.append([slope, intercept])
    
    
    # use kmean for clustering the lines
    kmean_model = KMeans(n_clusters=2, init = 'random', n_init=30).fit(np.array(line_params).reshape(-1,2))
    
    
    # extract average slope and intercept for each line
    line1_slope = kmean_model.cluster_centers_[0][0]
    line1_intercept = kmean_model.cluster_centers_[0][1]
    line2_slope = kmean_model.cluster_centers_[1][0]
    line2_intercept = kmean_model.cluster_centers_[1][1]
    
    
    # compute the distance from each endpoint to average line 1 and 2
    dist_to_line1 = []
    dist_to_line2 = []
    i=0
    for line in lines:
        for x1,y1,x2,y2 in line:
            if abs(x2 - x1) > 2:
                slope = (y2 - y1) / (x2 - x1)
                if abs(slope) > 0.5:
                    if kmean_model.labels_[i] == 0:
                        dist_to_line1.append(dist_to_line(x1, y1, line1_slope, line1_intercept))
                        dist_to_line1.append(dist_to_line(x2, y2, line1_slope, line1_intercept))
                    else:
                        dist_to_line2.append(dist_to_line(x1, y1, line2_slope, line2_intercept))
                        dist_to_line2.append(dist_to_line(x2, y2, line2_slope, line2_intercept))
                    i += 1
                
    # extract median and standard deviation from the distances
    mean_dist_to_line1 = np.median(np.array(dist_to_line1))
    mean_dist_to_line2 = np.median(np.array(dist_to_line2))
    std_dist_to_line1 = np.std(np.array(dist_to_line1))
    std_dist_to_line2 = np.std(np.array(dist_to_line2))
    
    
    # remove lines which are clearly outliers based on the distance from its endpoints to the average line
    outliers = [False for i in range(len(line_params))]
    i=0
    for line in lines:
        for x1,y1,x2,y2 in line:
            if abs(x2 - x1) > 2:
                slope = (y2 - y1) / (x2 - x1)
                if abs(slope) > 0.5:
                    if kmean_model.labels_[i] == 0:
                        if (dist_to_line(x1, y1, line1_slope, line1_intercept) > mean_dist_to_line1 + 1.5 * std_dist_to_line1 + 0.1) \
                            or (dist_to_line(x2, y2, line1_slope, line1_intercept) > mean_dist_to_line1 + 1.5 * std_dist_to_line1 + 0.1):
                            outliers[i] = True
                            #print("eliminate1 "+str(i))
                    else:
                        if (dist_to_line(x1, y1, line2_slope, line2_intercept) > mean_dist_to_line2 + 1.5 * std_dist_to_line2 + 0.1) \
                            or (dist_to_line(x2, y2, line2_slope, line2_intercept) > mean_dist_to_line2 + 1.5 * std_dist_to_line2 + 0.1):
                            outliers[i] = True
                            #print("eliminate2 "+str(i))
                    i += 1
    
    
    # recompute the slope and intercept of the average line1&2 after removing outliers
    line1_slope = 0.0
    line2_slope = 0.0
    line1_intercept = 0.0
    line2_intercept = 0.0
    num1 = 0
    num2 = 0
    i=0
    for line in lines:
        for x1,y1,x2,y2 in line:
            if abs(x2 - x1) > 2:
                slope = (y2 - y1) / (x2 - x1)
                if abs(slope) > 0.5:
                    if not outliers[i]:
                        if kmean_model.labels_[i] == 0:
                            line1_slope += line_params[i][0]
                            line1_intercept += line_params[i][1]
                            num1 += 1
                        else:
                            line2_slope += line_params[i][0]
                            line2_intercept += line_params[i][1]
                            num2 += 1
                    i += 1
    
    line1_intercept /= num1
    line2_intercept /= num2
    line1_slope /= num1
    line2_slope /= num2

    w = img.shape[1]
    h = img.shape[0]
    
    # both lines start from the bottom
    line1_lower_y = h
    line2_lower_y = h
    line1_lower_x = int((line1_lower_y - line1_intercept) / line1_slope)
    line2_lower_x = int((line2_lower_y - line2_intercept) / line2_slope)
    
    # both lines ends at the maximum y    
    line1_upper_y = int(0.60 * h)
    line2_upper_y = int(0.60 * h)
    line1_upper_x = int((line1_upper_y - line1_intercept) / line1_slope)
    line2_upper_x = int((line2_upper_y - line2_intercept) / line2_slope)
    
    """
    i=0
    for line in lines:
        for x1,y1,x2,y2 in line:
            if abs(x2 - x1) > 2:
                slope = (y2 - y1) / (x2 - x1)
                if abs(slope) > 0.5:
                    if kmean_model.labels_[i] == 0:
                        cv2.line(img, (x1, y1), (x2, y2), [0,255,0], thickness)
                    else:
                        cv2.line(img, (x1, y1), (x2, y2), [0,0,255], thickness)
                    i += 1
    """
    
    cv2.line(img, (line1_lower_x, line1_lower_y), (line1_upper_x, line1_upper_y), color, thickness)
    cv2.line(img, (line2_lower_x, line2_lower_y), (line2_upper_x, line2_upper_y), color, thickness)
                    
def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines, thickness=8)
    return line_img


def hough_transform_and_skectch_lanes(image, image_after_lane_region_masked):

    hough_line_img = hough_lines(image_after_lane_region_masked, rho=2, theta=np.pi/180, 
                                     threshold=20, min_line_len=30, max_line_gap=15)
    
    return weighted_img(hough_line_img, image)


processed_images = [hough_transform_and_skectch_lanes(test_images[i], images_after_canny_mask[i]) for i in range(len(test_images))]
show_processed_images(test_images, processed_images)

#test_images = [plt.imread(image_file) for image_file in glob.glob('test_images/*.jpg')]

i = 0
for image_file in glob.glob('test_images/*.jpg'):
    arr = re.split("[\\\.]+", image_file)
    #cv2.imwrite("test_images_output/"+arr[-2]+"_processed.jpg", processed_images[i])
    img = Image.fromarray(processed_images[i], 'RGB')
    img.save("test_images_output/"+arr[-2]+"_processed.jpg")
    i += 1
````
